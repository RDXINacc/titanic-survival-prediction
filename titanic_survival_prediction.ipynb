```python
import pandas as pd
train = pd.read_csv("train.csv")
train.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
train
```

```python
train_inputs = train.drop('Survived',axis='columns')
train_target = train.Survived
```

```python
train_inputs.Sex = train_inputs.Sex.map({'male': 0, 'female': 1})
```

```python
train_inputs.Age = train_inputs.Age.fillna(train_inputs.Age.mean())
```

```python
from sklearn.ensemble import RandomForestClassifier as rfc
model = rfc(n_estimators=10000, random_state=42)
```

```python
model.fit(train_inputs,train_target)
```

```python
test = pd.read_csv("test.csv")
test1 = test.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
test1
```

```python
test1.Sex = test1.Sex.map({'male': 0, 'female': 1})
```

```python
test1.Age = test1.Age.fillna(test1.Age.mean())
```

```python
y_predict = model.predict(test1)
y_predict
```

```python
model.score(test1,y_predict)
```

```python
test2 = test.drop(['Pclass','Sex','Age','Fare','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
test2
```

```python
test2['Survived']=y_predict
test2
```

```python
test2.to_csv('prediction_submission_PId_Survived.csv', index=False)
```
